```
import subprocess

with open('requirements.txt') as f:
  lines = f.readlines()

for line in lines:
  package = line.strip()
  result = subprocess.run(['pip', 'install', package], stdout=subprocess.PIPE, stderr=subprocess.STDOUT)
  print(result.stdout.decode())
```
